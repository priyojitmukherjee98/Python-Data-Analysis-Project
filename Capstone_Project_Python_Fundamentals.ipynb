```

# ================================
# TASK 1: CREATE DATAFRAMES & SAVE AS CSV
# ================================

# Import required libraries
import pandas as pd  # Import pandas for data handling
import numpy as np  # Import numpy for numerical operations

# Creating Employee DataFrame
employee_df = pd.DataFrame({  # Creating employee dataframe
    "ID": ["A001", "A002", "A003", "A004", "A005"],  # Employee IDs
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],  # Names
    "Gender": ["M", "F", "M", "F", "F"],  # Gender
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],  # City
    "Age": [25, 27, 29, 31, 30]  # Age
})

# Creating Seniority Level DataFrame
seniority_df = pd.DataFrame({  # Creating seniority dataframe
    "ID": ["A001", "A002", "A003", "A004", "A005"],  # Employee IDs
    "Designation Level": [2, 2, 3, 2, 3]  # Designation levels
})

# Creating Project DataFrame with missing cost values
project_df = pd.DataFrame({  # Creating project dataframe
    "ID": ["A001", "A002", "A003", "A004", "A005", "A002", "A005",
           "A003", "A001", "A003", "A001", "A004", "A004", "A005"],  # IDs
    "Project": ["Project 1", "Project 2", "Project 3", "Project 4", "Project 5",
                "Project 6", "Project 7", "Project 8", "Project 9",
                "Project 10", "Project 11", "Project 12", "Project 13", "Project 14"],  # Project names
    "Cost": [1002000, 2000000, 4500000, 5500000, np.nan,
             680000, 400000, 350000, np.nan,
             300000, 2000000, 1000000, 3000000, 200000],  # Cost values
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished",
               "Failed", "Finished", "Failed", "Ongoing",
               "Finished", "Failed", "Ongoing", "Finished", "Finished"]  # Status
})

# Save DataFrames as CSV
employee_df.to_csv("Employee.csv", index=False)  # Save employee CSV
seniority_df.to_csv("Seniority.csv", index=False)  # Save seniority CSV
project_df.to_csv("Project.csv", index=False)  # Save project CSV


# ================================
# TASK 2: FILL MISSING COST USING RUNNING AVERAGE (FOR LOOP)
# ================================

project_df = pd.read_csv("Project.csv")  # Read project CSV

running_sum = 0  # Initialize running sum
count = 0  # Initialize counter

for i in range(len(project_df)):  # Loop through rows
    if not pd.isna(project_df.loc[i, "Cost"]):  # If cost is not NaN
        running_sum += project_df.loc[i, "Cost"]  # Add to running sum
        count += 1  # Increase count
    else:
        project_df.loc[i, "Cost"] = running_sum / count  # Replace with running average

project_df.to_csv("Project.csv", index=False)  # Save updated CSV


# ================================
# TASK 3: SPLIT NAME INTO FIRST & LAST NAME
# ================================

employee_df = pd.read_csv("Employee.csv")  # Read employee CSV

employee_df[["First Name", "Last Name"]] = employee_df["Name"].str.split(" ", expand=True)  # Split name

employee_df.drop("Name", axis=1, inplace=True)  # Drop old name column

employee_df.to_csv("Employee.csv", index=False)  # Save updated CSV


# ================================
# TASK 4: MERGE ALL DATAFRAMES
# ================================

employee_df = pd.read_csv("Employee.csv")  # Read employee CSV
seniority_df = pd.read_csv("Seniority.csv")  # Read seniority CSV
project_df = pd.read_csv("Project.csv")  # Read project CSV

Final = pd.merge(employee_df, seniority_df, on="ID")  # Merge employee and seniority
Final = pd.merge(Final, project_df, on="ID")  # Merge with project


# ================================
# TASK 5: ADD BONUS COLUMN
# ================================

Final["Bonus"] = np.where(Final["Status"] == "Finished",
                          Final["Cost"] * 0.05,
                          0)  # 5% bonus if finished


# ================================
# TASK 6: DEMOTE FAILED & REMOVE ABOVE 4
# ================================

Final.loc[Final["Status"] == "Failed", "Designation Level"] += 1  # Demote

Final = Final[Final["Designation Level"] <= 4]  # Remove above 4


# ================================
# TASK 7: ADD MR/MRS & DROP GENDER
# ================================

Final["First Name"] = np.where(Final["Gender"] == "M",
                               "Mr. " + Final["First Name"],
                               "Mrs. " + Final["First Name"])  # Add prefix

Final.drop("Gender", axis=1, inplace=True)  # Drop gender


# ================================
# TASK 8: PROMOTE AGE > 29
# ================================

Final.loc[Final["Age"] > 29, "Designation Level"] -= 1  # Promote


# ================================
# TASK 9: TOTAL PROJECT COST PER EMPLOYEE
# ================================

TotalProjCost = Final.groupby(["ID", "First Name"])["Cost"].sum().reset_index()  # Group and sum

TotalProjCost.rename(columns={"Cost": "Total Cost"}, inplace=True)  # Rename column

print(TotalProjCost)  # Display result


# ================================
# TASK 10: PRINT CITY CONTAINS 'o'
# ================================

result = Final[Final["City"].str.contains("o", case=False)]  # Filter cities containing 'o'

print(result)  # Display result
```
